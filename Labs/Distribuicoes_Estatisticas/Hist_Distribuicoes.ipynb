## Exercício
Agora que você aprendeu os principais tipos de distribuições, o objetivo é utilizar a base de dados do censo para identificar as distribuições para cada um dos atributos. Para isso, carregue o arquivo census.csv e gere o histograma para cada atributo

```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
from scipy import stats
```

```python
# Base de Dados
df = pd.read_csv('../../populacao_e_amostra/census.csv')
df.head()
```

```python
# Tamanho
df.shape
```

```python
# Informação dos valores numéricos 
df.describe()
```

```python
# Gráfico
sns.histplot(df['age'], kde=True);
```

```python
work = df.groupby(['workclass']).size()
work.plot.barh()
```

```python
edu = df.groupby(['education']).size()
edu.plot.barh()
```

```python
numedu = df.groupby(['education-num']).size()
numedu.plot.barh()
```

```python
marital = df.groupby(['marital-status']).size()
marital.plot.barh()	
```

```python
occupation = df.groupby(['occupation']).size()
occupation.plot.barh()
```
