## Exercício Correlação

### Importações

```python
# Bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Base de Dados

```python
# Dados
ds = pd.read_csv('../../Distribuições_Estatisticas/house_prices.csv')
ds.head()
```

```python
# Visualizando os valores nulos
ds.isnull().sum()
```

```python
# Eliminado a coluna id e date
ds = ds.drop(['id', 'date'], axis=1)
ds.head()
```

```python
# Calculando a Covariância com Pandas
ds.cov()
```

```python
# Coeficiente de Correlação com Pandas
ds.corr()
```

### Resolução com o tutor

```python
# Eliminado a coluna id e date
ds = ds.drop(['sqft_living', 'sqft_lot'], axis=1)
ds.head()
```

```python
# Coeficiente de Correlação com Pandas
ds.corr()
```

```python
# Gráfico de correlação entre as variáveis sqft_living15, price
sns.scatterplot(x=ds['sqft_living15'], y=ds['price']);
```

```python
# Gráfico de correlação entre as variáveis Grade, price
sns.scatterplot(x=ds['grade'], y=ds['price']);
```

```python
# Gráfico de correlação entre as variáveis long, price
sns.scatterplot(x=ds['long'], y=ds['price']);
```

```python
# Heatmap
fig, ax = plt.subplots(figsize=(15, 15))
ax = sns.heatmap(ds.corr(), annot=True)
```
