### Exercício de Distribuição de Frequência
O objetivo desta tarefa é aplicar a distribuição de frequência utilizando o atributo age da base de dados do censo

- Carregue a base de dados census.csv
- Faça testes utilizando o parâmetro bins para visualizar a distribuição dos dados

```python
# Bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Base de dado
df = pd.read_csv('./../../populacao_e_amostra/census.csv')
df.head()
```

```python
# extraindo só a coluna idade
df = df.iloc[:, 0]
df.head()
```

```python
# Criando a frequência e o intervalo de classes
frequencia, classes = np.histogram(df)
frequencia, classes, len(classes)
```

```python
# Visualizando o gráfico
plt.hist(df, bins=classes);
```

```python
# gráfico com bins=6
frequencia, classes = np.histogram(df, bins=8)
frequencia, classes
```

```python
# Gráfico
plt.hist(df, classes);
```

```python
# Criando o dataset
ds = pd.DataFrame({'dados': df})
ds.head()
```

```python
# Gráfico
ds.plot.hist();
```

```python
# Gráfico com seaborn
sns.histplot(ds, kde=True);
```
