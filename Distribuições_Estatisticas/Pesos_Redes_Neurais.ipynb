## Inicialização de Pesos em Redes Neurais Artificiais
#### Inicializadores
- https://keras.io/api/layers/initializers/

```python
# Bibliotecas
import math
import numpy as np
import seaborn as sns
import tensorflow as tf
tf.__version__
```

```python
# Lib Inicializadores
from tensorflow.keras import initializers
```

### Random normal

```python
# Gerando dados normais e média, desvio padrão
normal = initializers.RandomNormal()
dados_normal = normal(shape=[1000])
np.mean(dados_normal), np.std(dados_normal)
```

```python
# Gráfico de dados
sns.histplot(dados_normal, kde=True);
```

### Random uniform

```python
# Dados Uniforme
uniforme = initializers.RandomUniform()
dados_uniforme = uniforme(shape=[1000])
# Gráfico
sns.histplot(dados_uniforme, kde=True);
```

### Glorot normal (Xavier initialization)
- Centered on 0 with stddev = sqrt(2 / (fan_in + fan_out)) where fan_in is the number of input units in the weight tensor and fan_out is the number of output units in the weight tensor
- https://keras.io/api/layers/initializers/

```python
# O desvio padrão
"""
10 -> 100 -> 100 -> 1
"""
math.sqrt(2 / (10 + 100)), math.sqrt(2 / (100 + 100)), math.sqrt(2 / (100 + 1))
```

```python
# Gerando dados glorot
normal_glorot = initializers.GlorotNormal()
dados_normal_glorot = normal_glorot(shape=[1000])
# Gráfico
sns.histplot(dados_normal_glorot, kde=True)
```

### Glorot Uniform

```python
# Dados glorot uniform
uniforme_glorot = initializers.GlorotUniform()
dados_uniforme_glorot = uniforme_glorot(shape=[1000])
# Gráfico
sns.histplot(dados_uniforme_glorot, kde=True)
```
